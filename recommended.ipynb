```python
import ast #string indisies must be integer
```

```python
def convert(obj): #beacuse of this loop we will get dict
    L = [] #empty loop to store
    for i in ast.literal_eval(obj):
        L.append(i['name'])  #printing only the name
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert) #getting a list of genures of that movie
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert) 
movies.head()
```

```python
import ast
ast.literal_eval('[{"movie_id": 28, "name": "Action"}, {"movie_id": 12, "name": "Adventure"}, {"movie_id": 14, "name": "Fantasy"}, {"movie_id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:    #only giving a count of 3 main character names
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):   #only to extract director name
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) 
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else []) 

```

```python
movies['overview'] = movies['overview'].apply(
    lambda x: x.split() if isinstance(x, str) else x if isinstance(x, list) else []
)

```

```python
#transforming the data
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x]) #trandforming data in between the space
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']] 
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x).lower() if isinstance(x, list) else x.lower()) #lower case
```

```python
#converting tags into vectors
#converting text to vector we have bag of word
import nltk  
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()  
```

```python
def stem(text):
    y = []
    for i in text.split():
      y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer #it is used to convert text to vector
cv = CountVectorizer(max_features=50000,stop_words='english')#5000 most frequent words
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #which used to convert numpy array

```

```python
vectors #convertion into vector form
```

```python
cv.get_feature_names_out()
```

```python
ps.stem('danced') #which is used to get similar words which used to give root word
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)  #similarity is less distance is more
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] 
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6] #fetching the 5 similarty movies
    for i in movies_list:
        print(new_df.iloc[i[0]].title) #provide movie title not the number
```

```python
recommend('Batman Begins')
```
